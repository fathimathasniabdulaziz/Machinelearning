```python
data.fillna(0, inplace=True)
data
```

```python
filtered_data = data[(data['Age'] > 40) & (data['Salary'] < 5000)]

print(filtered_data)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['Salary'], color='blue', alpha=0.5)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
place_counts = data['Place'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
place_counts.plot(kind='bar', color='skyblue')
plt.title('Number of People from Each Place')
plt.xlabel('Place')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#FEATURE SVALING
from sklearn.preprocessing import StandardScaler
```

```python
columns_to_scale = ['Age', 'Salary']  # Specify the columns you want to scale

# Standard Scaling
scaler = StandardScaler()
data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])

# Min-Max Scaling
# scaler = MinMaxScaler()
# df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

print(data)
```
